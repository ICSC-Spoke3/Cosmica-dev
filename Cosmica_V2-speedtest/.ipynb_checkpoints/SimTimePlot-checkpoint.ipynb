## Simulation launch configuration

- The number of 5000 particles is rounded to 5632 to fit the warpsize
- The number of warps per block was varied from 2 to 32 for each GPU board (only the avilable values of warps per block are taken into account)
- The number of blocks is computed by rounding the $\frac{Npart}{WpB*WarpSize}$ ratio
- The number of threads per block is computed by rounding the $\frac{Npart}{Nblocks}$ ratio

## Plot of the execution time in function of the warps per block

### Load data

```python
import numpy as np
import copy
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
import seaborn as sn

exe_times_dict = {
      "name": "",          # GPU used
      "WpB": np.array([]),
      "time": np.array([]) # seconds
}

code_version_list = ["WpB_speedtest_A30", "WpB_speedtest_A40"]
exe_times_dicts = np.array([])

for version_id in range(len(code_version_list)):
    try:
        exe_times_file = open(f"SimExeTimes_{code_version_list[version_id]}.txt")
        lines = exe_times_file.readlines()
        nSim = len(lines)
        local_exe_times_dict = copy.deepcopy(exe_times_dict)
        local_exe_times_dict["name"] = code_version_list[version_id].split("_")[-1]
        exe_times_file.close()

        for sim_id in range(nSim):
            line = lines[sim_id]
            exe_time = line.split()[11]
            WpB = int(line.split()[6])
            if exe_time != "X":
                local_exe_times_dict["time"] = np.append(local_exe_times_dict["time"], round(float(exe_time)/1000, 2))
                local_exe_times_dict["WpB"] = np.append(local_exe_times_dict["WpB"], WpB)

        exe_times_dicts = np.append(exe_times_dicts, local_exe_times_dict)

    except FileNotFoundError:
                    print(f"ERROR: SimExeTimes_{code_version_list[version_id]}.txt not found\n")
print(exe_times_dicts)
```

### Execution time comparison plot

```python
fig1, ax1 = plt.subplots()

min = np.inf
max = 0

# Fit the two data sample with possible functions
def func1(x, a, b, c):
    return a*(x-c)+b

def func2(x, a, b, c):
    return a*(x-c)**2 +b

fit_func_list = [{"name": "linear", "func": func1, "params": ["a", "b", "c"]},
                 {"name": "quadratic", "func": func2, "params": ["a", "b", "c"]}]


popts = np.empty((len(code_version_list), len(fit_func_list)))
pcovs = np.empty((len(code_version_list), len(fit_func_list)))

# Plot the A30 and A 40 extracted execution time
for version_id in range(len(code_version_list)):
    ax1.scatter(exe_times_dicts[version_id]["WpB"], exe_times_dicts[version_id]["time"], label=exe_times_dicts[version_id]["name"]+" HelMod", marker="o")
    for fit_func_id in range(len(fit_func_list)):
        function = fit_func_list[fit_func_id]
        popt, pcov = curve_fit(function["func"], exe_times_dicts[version_id]["WpB"], exe_times_dicts[version_id]["time"])
        # popts[version_id][fit_func_id] = {"params": popt}
        # pcovs[version_id][fit_func_id] = {"params": pcov}
        fit_name = function["name"]
        GPU_name = exe_times_dicts[version_id]["name"]
        ax1.plot(exe_times_dicts[version_id]["WpB"], function["func"](exe_times_dicts[version_id]["WpB"], *popt), '--', label=f'{fit_name} fit {GPU_name}')
        print(f"Best fitting parameters for {fit_name} function on {GPU_name} board:")
        for par_id in range(len(function["params"])):
            locsal_params = function["params"]
            print(f"{locsal_params[par_id]}={popt[par_id]:.2f}")

    max_ = np.max(exe_times_dicts[version_id]["time"])
    if max_>max:
        max = max_
    min_ = np.min(exe_times_dicts[version_id]["time"])
    if min_<min:
        min = min_
max_speedup = max/min

"""for version_id in range(len(code_version_list)):
    popt, pcov = curve_fit(func2, exe_times_dicts[version_id]["WpB"], exe_times_dicts[version_id]["time"])
    popts[1][version_id] = popt
    pcovs[1][version_id] = pcov
    name = exe_times_dicts[version_id]["name"]
    ax1.plot(exe_times_dicts[version_id]["WpB"], func2(exe_times_dicts[version_id]["WpB"], *popt), '--', label=f'{fit_func_list[0]} fit {name}')
    print(f"Best fitting parameters for quadratic function on {name} board: a={popt[0]:.2f}, b={popt[1]:.2f}, c={popt[2]:.2f}")
"""
# Show of the plot and set of its box
ax1.set_title("Execution time on different GPUs with respect warps per block")
ax1.set_xlabel("WpB")
ax1.set_ylabel("time (s)")
ax1.set_ylim(0, max+1)
ax2 = ax1.twinx()
tot_time = np.sort(np.append(exe_times_dicts[0]["time"], exe_times_dicts[1]["time"]))
speedup_scale = np.round(max/tot_time, 2)
f = lambda a: 1/(a+1e-8)
g = lambda b: 1/(b+1e-8)
plt.yscale("function", functions=(f, g))
ax2_values = np.arange(1, max_speedup+1, 0.5)
ax2.set_yticks(ax2_values)
ax2.set_ylim(max/(max+1), 1/1e-8)
# ax2.invert_yaxis()
ax2.set_ylabel("speedup")
ax1.legend()
ax1.grid()
fig1.savefig("SimExeTimes_compare.pdf")
fig1.show()

# Plot of the covariant matrix of the fitting parameters
fig2, ax3 = plt.subplots(2, 2)

for version_id in range(len(code_version_list)):
    for fit_func_id in range(len(fit_func_list)):
        cov_metrix = pd.DataFrame(pcovs[version_id][fit_func_id], index=fit_func_list[fit_func_id]["params"], columns=fit_func_list[fit_func_id]["params"])
        ax3[version_id][version_id] = sn.heatmap(cov_metrix, annot=True, fmt=".1f")
        fit_name = fit_func_list[fit_func_id]["name"]
        GPU_name = exe_times_dicts[version_id]["name"]
        ax3[version_id][version_id].set_title(f"{fit_name} function on {GPU_name} board")
        
fig2.suptitle("Covariance matix of the fitted functions\n")
fig2.show()
```
